```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import matplotlib as mpl
```

```python
df = pd.read_csv("data_mobil.csv")
reg = LinearRegression()
reg.fit(df[['x1','x2']].values,df.y)
```

```python
mpl.rcParams['legend.fontsize'] = 12
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(df.x1, df.x2, df.y)
ax.legend()
ax.view_init(45, 0)
plt.show()
```

```python
a = reg.coef_
b = reg.intercept_
```

```python
X_baru = np.array([900, 114]).reshape(-1, 2)
prediksi_Y = reg.predict(X_baru)
```

```python
print('Koefisien regresi = ' , a[0], ' dan ', a[1])
print('Intecept = ', b)
print('Hasil prediksi untuk x1=900 dan x2=114 yaitu ', prediksi_Y)
```
